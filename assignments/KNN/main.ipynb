## Diabetes Detector
#### KNN assignment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
xtrain = pd.read_csv('./data/train/Diabetes_XTrain.csv').values
ytrain = pd.read_csv('./data/train/Diabetes_YTrain.csv').values
ytrain = ytrain.reshape((-1,))
xtest = pd.read_csv('./data/test/Diabetes_Xtest.csv').values

print(xtrain.shape, ytrain.shape, xtest.shape)
```

```python
model = KNeighborsClassifier()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
diabetic = (ypred == 1).sum()
healthy = (ypred == 0).sum()
categories = ['Diabetic', 'Healthy']
values = [diabetic, healthy]
colors = ['red','green']

plt.bar(categories, values, color=colors)
plt.xlabel('Test Result')
plt.ylabel('Number of people')
plt.title('Diabetes Test')
plt.show()
```
