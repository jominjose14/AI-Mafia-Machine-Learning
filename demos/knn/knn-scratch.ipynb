## K Nearest Neighbors
#### Implemented from scratch for classification use case

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

```python
dfx = pd.read_csv('./data/xdata.csv')
dfy = pd.read_csv('./data/ydata.csv')

x=dfx.values
y=dfy.values

x=x[:,1:]
y=y[:,1:]

y=y.reshape((-1,))

print(x.shape, y.shape)
```

```python
query = np.array([2,3])
plt.scatter(query[0], query[1], c='red')
plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

```python
def dist(x1, x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(x,y,query,k=5):
    pairs = [] # pair(distance, class)
    
    for i in range(x.shape[0]):
        d=dist(query, x[i])
        pairs.append((d,y[i]))
        
    pairs = sorted(pairs)
    pairs = pairs[:k] # isolate first k pairs
    
    pairs = np.array(pairs)
    pair_counts = np.unique(pairs[:,1], return_counts=True)
    # print(pair_counts)
    idx = pair_counts[1].argmax()
    
    pred = pair_counts[0][idx]
    return pred
```

```python
knn(x,y,[2,3])
```
