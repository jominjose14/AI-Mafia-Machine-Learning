## K Means Clustering
##### implemented from scratch

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples=1000, n_features=2, centers=5, random_state=10)
```

```python
plt.scatter(x[:,0], x[:,1], c=y)
plt.show()
```

```python
k = 5
colors = ['green', 'blue', 'red', 'yellow', 'gray']
clusters = {}

for i in range(k):
    center = 10 * (2 * np.random.random((x.shape[1],)) - 1)
    points = []
    cluster = {
        'center': center,
        'points': points,
        'color': colors[i]
    }
    clusters[i] = cluster
```

```python
clusters
```

```python
def dist(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assign(clusters):
    for point in x:
        distances = []
        for i in range(k):
            d = dist(point, clusters[i]['center'])
            distances.append(d)
            
        clstr_idx = np.argmin(distances)
        clusters[clstr_idx]['points'].append(point)
```

```python
def updateCluster(clusters):
    for i in range(k):
        points = np.array(clusters[i]['points'])
        if 0 < points.shape[0]:
            new_center = points.mean(axis=0)
            clusters[i]['center'] = new_center
            clusters[i]['points'] = []
```

```python
def plot_clusters(clusters):
    for i in range(k):
        points = np.array(clusters[i]['points'])
        if 0 < points.shape[0]:
            plt.scatter(points[:,0], points[:,1], color=clusters[i]['color'])
            
        plt.scatter(clusters[i]['center'][0], clusters[i]['center'][1], color='black', marker='*')
```

```python
assign(clusters)
plot_clusters(clusters)
updateCluster(clusters)
```
