### Web Scraper

Considerations
- check sites terms and conditions
- avoid DOS attack, space out requests
- keep check on scraper (URLs may break, websites may go down)

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Fantastic_Four'
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
```

```python
content = soup.findAll('p')
print(content)
```
